```python
import ast
```

```python
def convert(obj):
    L = []
    for item in ast.literal_eval(obj):
        L.append(item["name"])
    return L
```

```python
def main_cast(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(main_cast) #Just took the main actor
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
def recommend(movie):
    try:
        movie_index = new_movies_df[new_movies_df['title']==movie].index[0]
    except IndexError:
        return "Sorry, the movie title that you requested does not appear to be in my database. Please try another movie!"
    
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:10]
    
    for i in movies_list:
        print(new_movies_df.iloc[i[0]].title)
        
```

#### Making a recommendation based off popularity

First of all we will need to create a data frame which contains only the movies with more than 100 votes, this is to ensure that the vote average has been balanced out. Those deep cuts which have 1 vote wont be represented in the data and wont be able to be recommended through this method.

```python
popular_movies_df = movies_df.copy()
popular_movies_df = popular_movies_df[popular_movies_df['vote_count'] >= 100]
popular_movies_df.sort_values('vote_average', ascending=False).head(50)
```

#### Allowing genre searching, and reciving feedback dependant on collective filtering

This function allows the user to filter by genres. It can take in one or plural genres, adding more genres will narrow the search as it needs to be match all given elements.

```python
def genre_recommendation(genres):
    if not isinstance(genres, list):
        genres = [genres]
    genre_movies = popular_movies_df.copy()
    genres = [genre.lower() for genre in genres]
    genre_movies = genre_movies[genre_movies['genres'].apply(lambda x: all([genre in [i.lower() if type(i) == str else i for i in x] for genre in genres]))]
    genre_movies = genre_movies.sort_values('vote_average', ascending=False).iloc[:10]
    
    movies_string = ''
    for i, row in enumerate(genre_movies.itertuples()):
        movie_string = f"{i+1}. {row.title} ({', '.join(row.genres)}) - {row.vote_average:.2f}\n"
        movies_string += movie_string
    
    genres_str = ', '.join(genres).capitalize()
    print(f"Top 10 recommended {genres_str} movies:\n{movies_string}")
```

```python
genre_recommendation(['action','adventure'])
```
